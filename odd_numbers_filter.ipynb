
## Завдання з об'єктно-орієнтованого програмування

Потрібно створити клас `OddNumbersFilter`, який працює зі списком чисел.

Вимоги:
- клас приймає список чисел у конструкторі
- реалізує метод для фільтрації непарних чисел
- має властивість (property), яка повертає список непарних чисел
- реалізує метод __str__ для виведення імені та групи студента
- продемонструвати роботу класу



### AI Prompt

Створи клас на Python з використанням об'єктно-орієнтованого програмування, який
приймає список чисел, фільтрує лише непарні числа, використовує метод і property,
а також реалізує метод __str__.


```

class OddNumbersFilter:
    STUDENT_NAME = "Максим"
    STUDENT_GROUP = "КНК-33"

    def __init__(self, numbers):
        self.numbers = numbers

    def filter(self):
        return [num for num in self.numbers if isinstance(num, int) and num % 2 != 0]

    @property
    def odd_numbers(self):
        return self.filter()

    def __str__(self):
        return f"Виконавець роботи: {self.STUDENT_NAME}\nГрупа: {self.STUDENT_GROUP}"


objects = [
    OddNumbersFilter([1, 2, 3, 4, 5, 6, 7]),
    OddNumbersFilter([10, 11, 12, 13, 14, 15]),
    OddNumbersFilter([21, 22, 23, 24, 25])
]

for obj in objects:
    print(obj)
    print("Вхідні дані:", obj.numbers)
    print("Непарні числа:", obj.odd_numbers)
    print("-" * 40)
```
